create csv file with the following columns:

Filename, Description, Keywords, Categories, Mature content, Editorial

```python
import os

# folder path
dir_path = r'C:\Users\yonat\Desktop\ShutterStock'

# list to store files
res = []
# Iterate directory
for file in os.listdir(dir_path):
    # check only text files
    if file.endswith('.jpg'):
        res.append(file)
print(res)
```

```python
import pandas as pd

columns = ['Filename', 'Description', 'Keywords', 'Categories', 'Mature content', 'Editorial']

df = pd.DataFrame(columns=columns)

df['Filename'] = res

df['Mature content'] = 'No'

df['Editorial'] = 'No'

df['Categories'] = 'Parks/Outdoor'
```

```python
df.to_csv('ShutterStock.csv', index=False)
df
```
