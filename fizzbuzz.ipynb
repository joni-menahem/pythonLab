multiply of 3 is fizz
multiply of 5 is buzz
multiply of both is fizzbuzz


```python
counter = 1
for i in range(100):
    if counter % 3 == 0 and counter % 5 == 0:
        print("fizzbuzz")
    elif counter % 3 == 0:
        print("fizz")
    elif counter % 5 == 0:
        print("buzz")
    else:
        print(counter)
    counter += 1
```
