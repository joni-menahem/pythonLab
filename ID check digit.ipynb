## ID check digit

```python
def is_id_valid(id):
    if len(id) != 9 or id == None:
        # print(len(id))
        # print("Make sure ID is formatted properly")
        return False
    sum = 0
    incNum = None
    for i in range(len(id)):
        incNum = int(id[i])*((i % 2) + 1) # Multiply number by 1 or 2
        # print(incNum, i)
        if incNum > 9:
            sum += incNum % 10 + 1
            incNum = None
        elif incNum <= 9:
            sum += incNum
    # print("sum:", sum)
    return (sum % 10 == 0)
```

```python
known_id = '8476'
counter = 0
for i in range(99999):
    id = str(i) + known_id
    bool = is_id_valid(id)
    if bool and (id[0] == '2' or id[0] == '3'):
        counter += 1
        print(id)

print(counter)
    
```
