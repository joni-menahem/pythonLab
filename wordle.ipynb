```python
letters = ['q','y','p','f','g','h','j','k','l','z','x']
```

```python
words = []
for first in range(0, len(letters)):
    for second in range(0, len(letters)):
        words.append('a' + letters[first] + 'o' + letters[second] + 't')  
```

```python
i = 0
for word in words:
    print('word number {0} is: {1}'.format(i, word))
    i += 1
```
